```python
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
#it will not work because of string so we will convert string of list into list
#so we use import ast 
# and inside ast there is function litral_eval from this we can convert string of list into list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# we want only top three cast of movie
from itertools import count
def convert(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3 :
      L.append(i['name'])
      counter = counter+1
    else:
      break
  return L
    
```

```python
movies['cast'] = movies['cast'].apply(convert)
#movies['crew'].apply(convert)
movies.head()
```

```python
#fetch director name from crew column
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job']== "Director":
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
# converting overview table in list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#trim the space(" "-->"") its help our machine will not going to confuse like siddharth maurya if name is give it will confuse who is siddhart and who is maurya
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
# porterstemmer helps in not cofusing about word like --> loving,loved --> it always give "love"
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
#cv.fit_transform(new_df['tags'],toarrays())
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[1])),reverse = True , key = lambda x:x[1])[0:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distance =similarity[movie_index]
  movie_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x: x[1])[1:6]
  for i in movie_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
